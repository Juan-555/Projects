```python
#funcion para cambiar el tipo de la columna con el fin de ver si existe relacion entre el director, la compania y lo que se recauda 
#function to see if the director, the company and the genre are relatated to movie revenue
df_correlation = Clean.copy()
for colum in df_correlation.columns:
    if(df_correlation[colum].dtype == "object" ):
        df_correlation[colum] = df_correlation[colum].astype("category")
        df_correlation[colum] = df_correlation[colum].cat.codes


df_correlation.drop(["country", "star", "writer"], axis=1, inplace=True)        
df_correlation.head() 
```
